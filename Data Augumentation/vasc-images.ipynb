```python
# original
vasc_images = ind_vasc[:]

# noise
vasc_noised = noise_image(vasc_images)


# flip
def flip_vasc_images(vasc_images):
    vasc_horizontal = flip_image(vasc_images, 'horizontal')
    vasc_vertical = flip_image(vasc_images, 'vertical')
    
    return vasc_horizontal + vasc_vertical

vasc_flipped = flip_vasc_images(vasc_images)


# rotated
def rotate_vasc_images(vasc_images):
    vasc_15 = rotate_image(vasc_images, 15)
    vasc_15_cw = rotate_image(vasc_images, 15, True)

    vasc_30 = rotate_image(vasc_images, 30)
    vasc_30_cw = rotate_image(vasc_images, 30, True)

    vasc_45 = rotate_image(vasc_images, 45)
    vasc_45_cw = rotate_image(vasc_images, 45, True)

    vasc_60 = rotate_image(vasc_images, 60)
    vasc_60_cw = rotate_image(vasc_images, 60, True)

    vasc_120 = rotate_image(vasc_images, 120)
    vasc_120_cw = rotate_image(vasc_images, 120, True)

    vasc_150 = rotate_image(vasc_images, 150)
    vasc_150_cw = rotate_image(vasc_images, 150, True)

    vasc_195 = rotate_image(vasc_images, 195)
    vasc_195_cw = rotate_image(vasc_images, 190, True)
    
    vasc_rotated = vasc_15 + vasc_30 + vasc_45 + vasc_60 + vasc_120 + vasc_150 + vasc_195
    vasc_rotated = vasc_rotated + vasc_15_cw + vasc_30_cw + vasc_45_cw + vasc_60_cw + vasc_120_cw + vasc_150_cw + vasc_195_cw

    return random.sample(vasc_rotated, 142*8)
    
vasc_rotated = rotate_vasc_images(vasc_images)
```

```python
# noise + flip
def flip_noised_vasc_images(vasc_noised):
    vasc_noised_horizontal = flip_image(vasc_noised, 'horizontal')
    vasc_noised_vertical = flip_image(vasc_noised, 'vertical')

    return vasc_noised_horizontal + vasc_noised_vertical

vasc_noised_flipped = flip_noised_vasc_images(vasc_noised)


# noise + rotate
def rotate_noised_vasc_images(vasc_noised):
    vasc_noised_15 = rotate_image(vasc_noised, 15)
    vasc_noised_15_cw = rotate_image(vasc_noised, 15, True)

    vasc_noised_30 = rotate_image(vasc_noised, 30)
    vasc_noised_30_cw = rotate_image(vasc_noised, 30, True)

    vasc_noised_45 = rotate_image(vasc_noised, 45)
    vasc_noised_45_cw = rotate_image(vasc_noised, 45, True)

    vasc_noised_135 = rotate_image(vasc_noised, 135)
    vasc_noised_135_cw = rotate_image(vasc_noised, 135, True)

    vasc_noised_165 = rotate_image(vasc_noised, 165)
    vasc_noised_165_cw = rotate_image(vasc_noised, 165, True)

    vasc_noised_210 = rotate_image(vasc_noised, 210)
    vasc_noised_210_cw = rotate_image(vasc_noised, 210, True)
    
    vasc_noised_rotated = vasc_noised_15 + vasc_noised_30 + vasc_noised_45 + vasc_noised_135 + vasc_noised_165 + vasc_noised_210
    vasc_noised_rotated = vasc_noised_rotated + vasc_noised_15_cw + vasc_noised_30_cw + vasc_noised_45_cw + vasc_noised_135_cw + vasc_noised_165_cw + vasc_noised_210_cw

    return random.sample(vasc_noised_rotated, 142*7)

vasc_noised_rotated = rotate_noised_vasc_images(vasc_noised)


# flip + rotate
def rotate_flipped_vasc_images(vasc_flipped):
    vasc_flipped_15 = rotate_image(vasc_flipped, 15)
    vasc_flipped_15_cw = rotate_image(vasc_flipped, 15, True)

    
    vasc_flipped_30 = rotate_image(vasc_flipped, 30)
    vasc_flipped_30_cw = rotate_image(vasc_flipped, 30, True)
    
    vasc_flipped_45 = rotate_image(vasc_flipped, 45)
    vasc_flipped_45_cw = rotate_image(vasc_flipped, 45, True)

    vasc_flipped_60 = rotate_image(vasc_flipped, 60)
    vasc_flipped_60_cw = rotate_image(vasc_flipped, 60, True)

    vasc_flipped_150 = rotate_image(vasc_flipped, 150)
    vasc_flipped_150_cw = rotate_image(vasc_flipped, 150, True)

    vasc_flipped_195 = rotate_image(vasc_flipped, 195)
    vasc_flipped_195_cw = rotate_image(vasc_flipped, 195, True)

    vasc_flipped_210 = rotate_image(vasc_flipped, 120)
    vasc_flipped_210_cw = rotate_image(vasc_flipped, 10, True)
    
    vasc_flipped_rotated = vasc_flipped_15 + vasc_flipped_30 + vasc_flipped_45 + vasc_flipped_60 + vasc_flipped_150 + vasc_flipped_195 + vasc_flipped_210
    vasc_flipped_rotated = vasc_flipped_rotated + vasc_flipped_15_cw + vasc_flipped_30_cw + vasc_flipped_45_cw + vasc_flipped_60_cw + vasc_flipped_150_cw + vasc_flipped_195_cw + vasc_flipped_210_cw
    
    return random.sample(vasc_flipped_rotated, 142*7)
    
vasc_flipped_rotated = rotate_flipped_vasc_images(vasc_flipped)
```

```python
# noise + flip + rotate
def rotate_noised_flipped_vasc_images(vasc_noised_flipped):
    vasc_noised_flipped_15 = rotate_image(vasc_noised_flipped, 15)
    vasc_noised_flipped_15_cw = rotate_image(vasc_noised_flipped, 15, True)

    vasc_noised_flipped_30 = rotate_image(vasc_noised_flipped, 30)
    vasc_noised_flipped_30_cw = rotate_image(vasc_noised_flipped, 30, True)

    vasc_noised_flipped_45 = rotate_image(vasc_noised_flipped, 45)
    vasc_noised_flipped_45_cw = rotate_image(vasc_noised_flipped, 45, True)

    vasc_noised_flipped_60 = rotate_image(vasc_noised_flipped, 60)
    vasc_noised_flipped_60_cw = rotate_image(vasc_noised_flipped, 60, True)

    vasc_noised_flipped_120 = rotate_image(vasc_noised_flipped, 120)
    vasc_noised_flipped_120_cw = rotate_image(vasc_noised_flipped, 120, True)

    vasc_noised_flipped_135 = rotate_image(vasc_noised_flipped, 135)
    vasc_noised_flipped_135_cw = rotate_image(vasc_noised_flipped, 135, True)

    vasc_noised_flipped_195 = rotate_image(vasc_noised_flipped, 195)
    vasc_noised_flipped_195_cw = rotate_image(vasc_noised_flipped, 195, True)

    vasc_noised_flipped_rotated = vasc_noised_flipped_15 + vasc_noised_flipped_30 + vasc_noised_flipped_45 + vasc_noised_flipped_60 + vasc_noised_flipped_120 + vasc_noised_flipped_135 + vasc_noised_flipped_195
    vasc_noised_flipped_rotated = vasc_noised_flipped_rotated + vasc_noised_flipped_15_cw + vasc_noised_flipped_30_cw + vasc_noised_flipped_45_cw + vasc_noised_flipped_60_cw + vasc_noised_flipped_120_cw + vasc_noised_flipped_135_cw + vasc_noised_flipped_195_cw

    return random.sample(vasc_noised_flipped_rotated, 142*8)

vasc_noised_flipped_rotated = rotate_noised_flipped_vasc_images(vasc_noised_flipped)
```
