```python
# original
akiec_images = ind_akiec[:]

# noise
akiec_noised = noise_image(akiec_images)


# flip
def flip_akiec_images(akiec_images):
    akiec_horizontal = flip_image(akiec_images, 'horizontal')
    akiec_vertical = flip_image(akiec_images, 'vertical')
    
    return akiec_horizontal + akiec_vertical

akiec_flipped = flip_akiec_images(akiec_images)


# rotated
def rotate_akiec_images(akiec_images):
    akiec_15 = rotate_image(akiec_images, 15)
    akiec_15_cw = rotate_image(akiec_images, 15, True)

    akiec_30 = rotate_image(akiec_images, 30)
    akiec_30_cw = rotate_image(akiec_images, 30, True)

    akiec_45 = rotate_image(akiec_images, 45)
    akiec_45_cw = rotate_image(akiec_images, 45, True)

    akiec_60 = rotate_image(akiec_images, 60)
    akiec_60_cw = rotate_image(akiec_images, 60, True)

    akiec_120 = rotate_image(akiec_images, 120)
    akiec_120_cw = rotate_image(akiec_images, 120, True)

    akiec_150 = rotate_image(akiec_images, 150)
    akiec_150_cw = rotate_image(akiec_images, 150, True)

    akiec_195 = rotate_image(akiec_images, 195)
    akiec_195_cw = rotate_image(akiec_images, 190, True)
    
    akiec_rotated = akiec_15 + akiec_30 + akiec_45 + akiec_60 + akiec_120 + akiec_150 + akiec_195
    akiec_rotated = akiec_rotated + akiec_15_cw + akiec_30_cw + akiec_45_cw + akiec_60_cw + akiec_120_cw + akiec_150_cw + akiec_195_cw

    return random.sample(akiec_rotated, 327*4)
    
akiec_rotated = rotate_akiec_images(akiec_images)
```

```python
# noise + flip
def flip_noised_akiec_images(akiec_noised):
    akiec_noised_horizontal = flip_image(akiec_noised, 'horizontal')
    akiec_noised_vertical = flip_image(akiec_noised, 'vertical')

    return akiec_noised_horizontal + akiec_noised_vertical

akiec_noised_flipped = flip_noised_akiec_images(akiec_noised) # 2


# noise + rotate
def rotate_noised_akiec_images(akiec_noised):
    akiec_noised_15 = rotate_image(akiec_noised, 15)
    akiec_noised_15_cw = rotate_image(akiec_noised, 15, True)

    akiec_noised_30 = rotate_image(akiec_noised, 30)
    akiec_noised_30_cw = rotate_image(akiec_noised, 30, True)

    akiec_noised_45 = rotate_image(akiec_noised, 45)
    akiec_noised_45_cw = rotate_image(akiec_noised, 45, True)

    akiec_noised_135 = rotate_image(akiec_noised, 135)
    akiec_noised_135_cw = rotate_image(akiec_noised, 135, True)

    akiec_noised_165 = rotate_image(akiec_noised, 165)
    akiec_noised_165_cw = rotate_image(akiec_noised, 165, True)

    akiec_noised_210 = rotate_image(akiec_noised, 210)
    akiec_noised_210_cw = rotate_image(akiec_noised, 210, True)
    
    akiec_noised_rotated = akiec_noised_15 + akiec_noised_30 + akiec_noised_45 + akiec_noised_135 + akiec_noised_165 + akiec_noised_210
    akiec_noised_rotated = akiec_noised_rotated + akiec_noised_15_cw + akiec_noised_30_cw + akiec_noised_45_cw + akiec_noised_135_cw + akiec_noised_165_cw + akiec_noised_210_cw

    return random.sample(akiec_noised_rotated, 327*2) # 2

akiec_noised_rotated = rotate_noised_akiec_images(akiec_noised)


# flip + rotate
def rotate_flipped_akiec_images(akiec_flipped):
    akiec_flipped_15 = rotate_image(akiec_flipped, 15)
    akiec_flipped_15_cw = rotate_image(akiec_flipped, 15, True)

    
    akiec_flipped_30 = rotate_image(akiec_flipped, 30)
    akiec_flipped_30_cw = rotate_image(akiec_flipped, 30, True)
    
    akiec_flipped_45 = rotate_image(akiec_flipped, 45)
    akiec_flipped_45_cw = rotate_image(akiec_flipped, 45, True)

    akiec_flipped_60 = rotate_image(akiec_flipped, 60)
    akiec_flipped_60_cw = rotate_image(akiec_flipped, 60, True)

    akiec_flipped_150 = rotate_image(akiec_flipped, 150)
    akiec_flipped_150_cw = rotate_image(akiec_flipped, 150, True)

    akiec_flipped_195 = rotate_image(akiec_flipped, 195)
    akiec_flipped_195_cw = rotate_image(akiec_flipped, 195, True)

    akiec_flipped_210 = rotate_image(akiec_flipped, 120)
    akiec_flipped_210_cw = rotate_image(akiec_flipped, 10, True)
    
    akiec_flipped_rotated = akiec_flipped_15 + akiec_flipped_30 + akiec_flipped_45 + akiec_flipped_60 + akiec_flipped_150 + akiec_flipped_195 + akiec_flipped_210
    akiec_flipped_rotated = akiec_flipped_rotated + akiec_flipped_15_cw + akiec_flipped_30_cw + akiec_flipped_45_cw + akiec_flipped_60_cw + akiec_flipped_150_cw + akiec_flipped_195_cw + akiec_flipped_210_cw
    
    return random.sample(akiec_flipped_rotated, 327*2) # 2
    
akiec_flipped_rotated = rotate_flipped_akiec_images(akiec_flipped) # 2
```

```python
# noise + flip + rotate
def rotate_noised_flipped_akiec_images(akiec_noised_flipped):
    akiec_noised_flipped_15 = rotate_image(akiec_noised_flipped, 15)
    akiec_noised_flipped_15_cw = rotate_image(akiec_noised_flipped, 15, True)

    akiec_noised_flipped_30 = rotate_image(akiec_noised_flipped, 30)
    akiec_noised_flipped_30_cw = rotate_image(akiec_noised_flipped, 30, True)

    akiec_noised_flipped_45 = rotate_image(akiec_noised_flipped, 45)
    akiec_noised_flipped_45_cw = rotate_image(akiec_noised_flipped, 45, True)

    akiec_noised_flipped_60 = rotate_image(akiec_noised_flipped, 60)
    akiec_noised_flipped_60_cw = rotate_image(akiec_noised_flipped, 60, True)

    akiec_noised_flipped_120 = rotate_image(akiec_noised_flipped, 120)
    akiec_noised_flipped_120_cw = rotate_image(akiec_noised_flipped, 120, True)

    akiec_noised_flipped_135 = rotate_image(akiec_noised_flipped, 135)
    akiec_noised_flipped_135_cw = rotate_image(akiec_noised_flipped, 135, True)

    akiec_noised_flipped_195 = rotate_image(akiec_noised_flipped, 195)
    akiec_noised_flipped_195_cw = rotate_image(akiec_noised_flipped, 195, True)

    akiec_noised_flipped_rotated = akiec_noised_flipped_15 + akiec_noised_flipped_30 + akiec_noised_flipped_45 + akiec_noised_flipped_60 + akiec_noised_flipped_120 + akiec_noised_flipped_135 + akiec_noised_flipped_195
    akiec_noised_flipped_rotated = akiec_noised_flipped_rotated + akiec_noised_flipped_15_cw + akiec_noised_flipped_30_cw + akiec_noised_flipped_45_cw + akiec_noised_flipped_60_cw + akiec_noised_flipped_120_cw + akiec_noised_flipped_135_cw + akiec_noised_flipped_195_cw

    return random.sample(akiec_noised_flipped_rotated, 327*2) # 2

akiec_noised_flipped_rotated = rotate_noised_flipped_akiec_images(akiec_noised_flipped)
```
