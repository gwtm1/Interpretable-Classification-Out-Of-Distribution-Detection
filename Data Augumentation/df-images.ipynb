```python
# original
df_images = ind_df[:]

# noise
df_noised = noise_image(df_images)


# flip
def flip_df_images(df_images):
    df_horizontal = flip_image(df_images, 'horizontal')
    df_vertical = flip_image(df_images, 'vertical')
    
    return df_horizontal + df_vertical

df_flipped = flip_df_images(df_images)


# rotated
def rotate_df_images(df_images):
    df_15 = rotate_image(df_images, 15)
    df_15_cw = rotate_image(df_images, 15, True)

    df_30 = rotate_image(df_images, 30)
    df_30_cw = rotate_image(df_images, 30, True)

    df_45 = rotate_image(df_images, 45)
    df_45_cw = rotate_image(df_images, 45, True)

    df_60 = rotate_image(df_images, 60)
    df_60_cw = rotate_image(df_images, 60, True)

    df_120 = rotate_image(df_images, 120)
    df_120_cw = rotate_image(df_images, 120, True)

    df_150 = rotate_image(df_images, 150)
    df_150_cw = rotate_image(df_images, 150, True)

    df_195 = rotate_image(df_images, 195)
    df_195_cw = rotate_image(df_images, 190, True)
    
    df_rotated = df_15 + df_30 + df_45 + df_60 + df_120 + df_150 + df_195
    df_rotated = df_rotated + df_15_cw + df_30_cw + df_45_cw + df_60_cw + df_120_cw + df_150_cw + df_195_cw

    return random.sample(df_rotated, 115*10)
    
df_rotated = rotate_df_images(df_images)
```

```python
# noise + flip
def flip_noised_df_images(df_noised):
    df_noised_horizontal = flip_image(df_noised, 'horizontal')
    df_noised_vertical = flip_image(df_noised, 'vertical')

    return df_noised_horizontal + df_noised_vertical

df_noised_flipped = flip_noised_df_images(df_noised)


# noise + rotate
def rotate_noised_df_images(df_noised):
    df_noised_15 = rotate_image(df_noised, 15)
    df_noised_15_cw = rotate_image(df_noised, 15, True)

    df_noised_30 = rotate_image(df_noised, 30)
    df_noised_30_cw = rotate_image(df_noised, 30, True)

    df_noised_45 = rotate_image(df_noised, 45)
    df_noised_45_cw = rotate_image(df_noised, 45, True)

    df_noised_135 = rotate_image(df_noised, 135)
    df_noised_135_cw = rotate_image(df_noised, 135, True)

    df_noised_165 = rotate_image(df_noised, 165)
    df_noised_165_cw = rotate_image(df_noised, 165, True)

    df_noised_210 = rotate_image(df_noised, 210)
    df_noised_210_cw = rotate_image(df_noised, 210, True)
    
    df_noised_rotated = df_noised_15 + df_noised_30 + df_noised_45 + df_noised_135 + df_noised_165 + df_noised_210
    df_noised_rotated = df_noised_rotated + df_noised_15_cw + df_noised_30_cw + df_noised_45_cw + df_noised_135_cw + df_noised_165_cw + df_noised_210_cw

    return random.sample(df_noised_rotated, 115*10)

df_noised_rotated = rotate_noised_df_images(df_noised)


# flip + rotate
def rotate_flipped_df_images(df_flipped):
    df_flipped_15 = rotate_image(df_flipped, 15)
    df_flipped_15_cw = rotate_image(df_flipped, 15, True)

    
    df_flipped_30 = rotate_image(df_flipped, 30)
    df_flipped_30_cw = rotate_image(df_flipped, 30, True)
    
    df_flipped_45 = rotate_image(df_flipped, 45)
    df_flipped_45_cw = rotate_image(df_flipped, 45, True)

    df_flipped_60 = rotate_image(df_flipped, 60)
    df_flipped_60_cw = rotate_image(df_flipped, 60, True)

    df_flipped_150 = rotate_image(df_flipped, 150)
    df_flipped_150_cw = rotate_image(df_flipped, 150, True)

    df_flipped_195 = rotate_image(df_flipped, 195)
    df_flipped_195_cw = rotate_image(df_flipped, 195, True)

    df_flipped_210 = rotate_image(df_flipped, 120)
    df_flipped_210_cw = rotate_image(df_flipped, 10, True)
    
    df_flipped_rotated = df_flipped_15 + df_flipped_30 + df_flipped_45 + df_flipped_60 + df_flipped_150 + df_flipped_195 + df_flipped_210
    df_flipped_rotated = df_flipped_rotated + df_flipped_15_cw + df_flipped_30_cw + df_flipped_45_cw + df_flipped_60_cw + df_flipped_150_cw + df_flipped_195_cw + df_flipped_210_cw
    
    return random.sample(df_flipped_rotated, 115*10)
    
df_flipped_rotated = rotate_flipped_df_images(df_flipped)
```

```python
# noise + flip + rotate
def rotate_noised_flipped_df_images(df_noised_flipped):
    df_noised_flipped_15 = rotate_image(df_noised_flipped, 15)
    df_noised_flipped_15_cw = rotate_image(df_noised_flipped, 15, True)

    df_noised_flipped_30 = rotate_image(df_noised_flipped, 30)
    df_noised_flipped_30_cw = rotate_image(df_noised_flipped, 30, True)

    df_noised_flipped_45 = rotate_image(df_noised_flipped, 45)
    df_noised_flipped_45_cw = rotate_image(df_noised_flipped, 45, True)

    df_noised_flipped_60 = rotate_image(df_noised_flipped, 60)
    df_noised_flipped_60_cw = rotate_image(df_noised_flipped, 60, True)

    df_noised_flipped_120 = rotate_image(df_noised_flipped, 120)
    df_noised_flipped_120_cw = rotate_image(df_noised_flipped, 120, True)

    df_noised_flipped_135 = rotate_image(df_noised_flipped, 135)
    df_noised_flipped_135_cw = rotate_image(df_noised_flipped, 135, True)

    df_noised_flipped_195 = rotate_image(df_noised_flipped, 195)
    df_noised_flipped_195_cw = rotate_image(df_noised_flipped, 195, True)

    df_noised_flipped_rotated = df_noised_flipped_15 + df_noised_flipped_30 + df_noised_flipped_45 + df_noised_flipped_60 + df_noised_flipped_120 + df_noised_flipped_135 + df_noised_flipped_195
    df_noised_flipped_rotated = df_noised_flipped_rotated + df_noised_flipped_15_cw + df_noised_flipped_30_cw + df_noised_flipped_45_cw + df_noised_flipped_60_cw + df_noised_flipped_120_cw + df_noised_flipped_135_cw + df_noised_flipped_195_cw

    return random.sample(df_noised_flipped_rotated, 115*9)

df_noised_flipped_rotated = rotate_noised_flipped_df_images(df_noised_flipped)
```
