```python
# original
bkl_images = ind_bkl[:]

# noise
bkl_noised = noise_image(bkl_images)

# flip
def flip_bkl_images(bkl_images):
    bkl_horizontal = flip_image(bkl_images, 'horizontal')
    bkl_vertical = flip_image(bkl_images, 'vertical')

    bkl_flipped = bkl_horizontal + bkl_vertical
    return random.sample(bkl_flipped, 825)


bkl_flipped = flip_bkl_images(bkl_images) # 0.75

# rotated
def rotate_bkl_images(bkl_images):
    bkl_15 = rotate_image(bkl_images, 15)
    bkl_15_cw = rotate_image(bkl_images, 15, True)

    bkl_45 = rotate_image(bkl_images, 45)
    bkl_45_cw = rotate_image(bkl_images, 45, True)

    bkl_60 = rotate_image(bkl_images, 60)
    bkl_60_cw = rotate_image(bkl_images, 60, True)

    bkl_120 = rotate_image(bkl_images, 120)
    bkl_120_cw = rotate_image(bkl_images, 120, True)

    bkl_150 = rotate_image(bkl_images, 150)
    bkl_150_cw = rotate_image(bkl_images, 150, True)

    bkl_rotated = bkl_15 + bkl_45 + bkl_60 + bkl_120 + bkl_150 + bkl_15_cw + bkl_45_cw + bkl_60_cw + bkl_120_cw + bkl_150_cw
    return random.sample(bkl_rotated, 825)

bkl_rotated = rotate_bkl_images(bkl_images) # 0.75
```

```python
# noise + flip
def flip_noised_bkl_images(bkl_noised):
    bkl_noised_horizontal = flip_image(bkl_noised, 'horizontal')
    bkl_noised_vertical = flip_image(bkl_noised, 'vertical')
    
    bkl_noised_flipped = bkl_noised_horizontal + bkl_noised_vertical
    return random.sample(bkl_noised_flipped, 825)

bkl_noised_flipped = flip_noised_bkl_images(bkl_noised) # 0.75


# noise + rotate
def rotate_noised_bkl_images(bkl_noised):
    bkl_noised_10 = rotate_image(bkl_noised, 10)
    bkl_noised_10_cw = rotate_image(bkl_noised, 10, True)

    bkl_noised_30 = rotate_image(bkl_noised, 30)
    bkl_noised_30_cw = rotate_image(bkl_noised, 30, True)

    bkl_noised_135 = rotate_image(bkl_noised, 135)
    bkl_noised_135_cw = rotate_image(bkl_noised, 135, True)

    bkl_noised_165 = rotate_image(bkl_noised, 165)
    bkl_noised_165_cw = rotate_image(bkl_noised, 165, True)

    bkl_noised_210 = rotate_image(bkl_noised, 210)
    bkl_noised_210_cw = rotate_image(bkl_noised, 210, True)

    bkl_noised_rotated = bkl_noised_10 + bkl_noised_30 + bkl_noised_135 + bkl_noised_165 + bkl_noised_210 + bkl_noised_10_cw + bkl_noised_30_cw + bkl_noised_135_cw + bkl_noised_165_cw + bkl_noised_210_cw
    return random.sample(bkl_noised_rotated, 825)

bkl_noised_rotated = rotate_noised_bkl_images(bkl_noised) # 0.75


# flip + rotate
def rotate_flipped_bkl_images(bkl_flipped):
    bkl_flipped_15 = rotate_image(bkl_flipped, 15)
    bkl_flipped_15_cw = rotate_image(bkl_flipped, 15, True)

    bkl_flipped_45 = rotate_image(bkl_flipped, 450)
    bkl_flipped_45_cw = rotate_image(bkl_flipped, 45, True)

    bkl_flipped_150 = rotate_image(bkl_flipped, 150)
    bkl_flipped_150_cw = rotate_image(bkl_flipped, 150, True)

    bkl_flipped_195 = rotate_image(bkl_flipped, 195)
    bkl_flipped_195_cw = rotate_image(bkl_flipped, 195, True)

    bkl_flipped_210 = rotate_image(bkl_flipped, 120)
    bkl_flipped_210_cw = rotate_image(bkl_flipped, 10, True)

    bkl_flipped_rotated = bkl_flipped_15 + bkl_flipped_45 + bkl_flipped_150 + bkl_flipped_195 + bkl_flipped_210 + bkl_flipped_15_cw + bkl_flipped_45_cw + bkl_flipped_150_cw + bkl_flipped_195_cw + bkl_flipped_210_cw
    return random.sample(bkl_flipped_rotated, 825)

bkl_flipped_rotated = rotate_flipped_bkl_images(bkl_flipped) # 0.75
```

```python
# noise + flip + rotate
def rotate_noised_flipped_bkl_images(bkl_noised_flipped):
    bkl_noised_flipped_15 = rotate_image(bkl_noised_flipped, 15)
    bkl_noised_flipped_15_cw = rotate_image(bkl_noised_flipped, 15, True)

    bkl_noised_flipped_60 = rotate_image(bkl_noised_flipped, 60)
    bkl_noised_flipped_60_cw = rotate_image(bkl_noised_flipped, 60, True)

    bkl_noised_flipped_135 = rotate_image(bkl_noised_flipped, 135)
    bkl_noised_flipped_135_cw = rotate_image(bkl_noised_flipped, 135, True)

    bkl_noised_flipped_195 = rotate_image(bkl_noised_flipped, 195)
    bkl_noised_flipped_195_cw = rotate_image(bkl_noised_flipped, 195, True)
    
    bkl_noised_flipped_rotated = bkl_noised_flipped_15 + bkl_noised_flipped_60 + bkl_noised_flipped_135 + bkl_noised_flipped_195 + bkl_noised_flipped_15_cw + bkl_noised_flipped_60_cw + bkl_noised_flipped_135_cw + bkl_noised_flipped_195_cw
    
    return random.sample(bkl_noised_flipped_rotated, 825)

bkl_noised_flipped_rotated = rotate_noised_flipped_bkl_images(bkl_noised_flipped)  # 0.75
```
