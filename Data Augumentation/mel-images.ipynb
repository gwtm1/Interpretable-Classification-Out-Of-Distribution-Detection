```python
# original
mel_images = ind_mel[:]

# noise
mel_noised = noise_image(mel_images)

# flip
def flip_mel_images(mel_images):
    mel_horizontal = flip_image(mel_images, 'horizontal')
    mel_vertical = flip_image(mel_images, 'vertical')

    mel_flipped = mel_horizontal + mel_vertical

    return random.sample(mel_flipped, 835)


mel_flipped = flip_mel_images(mel_images) # 0.75

# rotated
def rotate_mel_images(mel_images):
    mel_15 = rotate_image(mel_images, 15)
    mel_15_cw = rotate_image(mel_images, 15, True)

    mel_30 = rotate_image(mel_images, 30)
    mel_30_cw = rotate_image(mel_images, 30, True)

    mel_45 = rotate_image(mel_images, 45)
    mel_45_cw = rotate_image(mel_images, 45, True)

    mel_120 = rotate_image(mel_images, 120)
    mel_120_cw = rotate_image(mel_images, 120, True)

    mel_195 = rotate_image(mel_images, 195)
    mel_195_cw = rotate_image(mel_images, 190, True)
    
    mel_rotated = mel_15 + mel_30 + mel_45 + mel_120 + mel_195 + mel_15_cw + mel_30_cw + mel_45_cw + mel_120_cw + mel_195_cw
    
    return random.sample(mel_rotated, 835)

mel_rotated = rotate_mel_images(mel_images) # 0.75
```

```python
# noise + flip
def flip_noised_mel_images(mel_noised):
    mel_noised_horizontal = flip_image(mel_noised, 'horizontal')
    mel_noised_vertical = flip_image(mel_noised, 'vertical')
    
    mel_noised_flipped = mel_noised_horizontal + mel_noised_vertical
    return random.sample(mel_noised_flipped, 835)

mel_noised_flipped = flip_noised_mel_images(mel_noised) # 0.75


# noise + rotate
def rotate_noised_mel_images(mel_noised):
    mel_noised_30 = rotate_image(mel_noised, 30)
    mel_noised_30_cw = rotate_image(mel_noised, 30, True)
    
    mel_noised_45 = rotate_image(mel_noised, 45)
    mel_noised_45_cw = rotate_image(mel_noised, 45, True)

    mel_noised_165 = rotate_image(mel_noised, 165)
    mel_noised_165_cw = rotate_image(mel_noised, 165, True)

    mel_noised_210 = rotate_image(mel_noised, 210)
    mel_noised_210_cw = rotate_image(mel_noised, 210, True)

    mel_noised_rotated =  mel_noised_30 + mel_noised_45 + mel_noised_165 + mel_noised_210 + mel_noised_30_cw + mel_noised_45_cw + mel_noised_165_cw + mel_noised_210_cw

    return random.sample(mel_noised_rotated, 835)

mel_noised_rotated = rotate_noised_mel_images(mel_noised) # 0.75


# flip + rotate
def rotate_flipped_mel_images(mel_flipped):
    mel_flipped_15 = rotate_image(mel_flipped, 15)
    mel_flipped_15_cw = rotate_image(mel_flipped, 15, True)

    mel_flipped_60 = rotate_image(mel_flipped, 60)
    mel_flipped_60_cw = rotate_image(mel_flipped, 60, True)

    mel_flipped_150 = rotate_image(mel_flipped, 150)
    mel_flipped_150_cw = rotate_image(mel_flipped, 150, True)

    mel_flipped_195 = rotate_image(mel_flipped, 195)
    mel_flipped_195_cw = rotate_image(mel_flipped, 195, True)

    mel_flipped_rotated =  mel_flipped_15 + mel_flipped_60 + mel_flipped_150 + mel_flipped_195 + mel_flipped_15_cw + mel_flipped_60_cw + mel_flipped_150_cw + mel_flipped_195_cw
    
    return random.sample(mel_flipped_rotated, 557) # 0.5

mel_flipped_rotated = rotate_flipped_mel_images(mel_flipped)
```

```python
# noise + flip + rotate
def rotate_noised_flipped_mel_images(mel_noised_flipped):
    mel_noised_flipped_15 = rotate_image(mel_noised_flipped, 15)
    mel_noised_flipped_15_cw = rotate_image(mel_noised_flipped, 15, True)

    mel_noised_flipped_30 = rotate_image(mel_noised_flipped, 30)
    mel_noised_flipped_30_cw = rotate_image(mel_noised_flipped, 30, True)

    mel_noised_flipped_120 = rotate_image(mel_noised_flipped, 120)
    mel_noised_flipped_120_cw = rotate_image(mel_noised_flipped, 120, True)

    mel_noised_flipped_135 = rotate_image(mel_noised_flipped, 135)
    mel_noised_flipped_135_cw = rotate_image(mel_noised_flipped, 135, True)

    mel_noised_flipped_rotated = mel_noised_flipped_15 + mel_noised_flipped_30 + mel_noised_flipped_120 + mel_noised_flipped_135 + mel_noised_flipped_15_cw + mel_noised_flipped_30_cw + mel_noised_flipped_120_cw + mel_noised_flipped_135_cw
    
    return random.sample(mel_flipped_rotated, 557) # 0.5

mel_noised_flipped_rotated = rotate_noised_flipped_mel_images(mel_noised_flipped)
```

```python
ind_mel = mel_images + mel_noised + mel_flipped + mel_rotated + mel_noised_flipped + mel_noised_rotated + mel_flipped_rotated + mel_noised_flipped_rotated
```
