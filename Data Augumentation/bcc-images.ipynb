```python
# original
bcc_images = ind_bcc[:]

# noise
bcc_noised = noise_image(bcc_images)


# flip
def flip_bcc_images(bcc_images):
    bcc_horizontal = flip_image(bcc_images, 'horizontal')
    bcc_vertical = flip_image(bcc_images, 'vertical')
    
    bcc_flipped = bcc_horizontal + bcc_vertical
    return random.sample(bcc_flipped, 771)

bcc_flipped = flip_bcc_images(bcc_images) # 1.5


# rotated
def rotate_bcc_images(bcc_images):
    bcc_15 = rotate_image(bcc_images, 15)
    bcc_15_cw = rotate_image(bcc_images, 15, True)

    bcc_30 = rotate_image(bcc_images, 30)
    bcc_30_cw = rotate_image(bcc_images, 30, True)

    bcc_45 = rotate_image(bcc_images, 45)
    bcc_45_cw = rotate_image(bcc_images, 45, True)

    bcc_60 = rotate_image(bcc_images, 60)
    bcc_60_cw = rotate_image(bcc_images, 60, True)

    bcc_120 = rotate_image(bcc_images, 120)
    bcc_120_cw = rotate_image(bcc_images, 120, True)

    bcc_150 = rotate_image(bcc_images, 150)
    bcc_150_cw = rotate_image(bcc_images, 150, True)

    bcc_195 = rotate_image(bcc_images, 195)
    bcc_195_cw = rotate_image(bcc_images, 190, True)
    
    bcc_rotated = bcc_15 + bcc_30 + bcc_45 + bcc_60 + bcc_120 + bcc_150 + bcc_195 + bcc_15_cw + bcc_30_cw + bcc_45_cw + bcc_60_cw + bcc_120_cw + bcc_150_cw + bcc_195_cw
    return random.sample(bcc_rotated, 771)
    
bcc_rotated = rotate_bcc_images(bcc_images) # 1.5
```

```python
# noise + flip
def flip_noised_bcc_images(bcc_noised):
    bcc_noised_horizontal = flip_image(bcc_noised, 'horizontal')
    bcc_noised_vertical = flip_image(bcc_noised, 'vertical')

    bcc_noised_flipped = bcc_noised_horizontal + bcc_noised_vertical
    return random.sample(bcc_noised_flipped, 514)

bcc_noised_flipped = flip_noised_bcc_images(bcc_noised) # 1


# noise + rotate
def rotate_noised_bcc_images(bcc_noised):
    bcc_noised_15 = rotate_image(bcc_noised, 15)
    bcc_noised_15_cw = rotate_image(bcc_noised, 15, True)

    bcc_noised_30 = rotate_image(bcc_noised, 30)
    bcc_noised_30_cw = rotate_image(bcc_noised, 30, True)

    bcc_noised_45 = rotate_image(bcc_noised, 45)
    bcc_noised_45_cw = rotate_image(bcc_noised, 45, True)

    bcc_noised_135 = rotate_image(bcc_noised, 135)
    bcc_noised_135_cw = rotate_image(bcc_noised, 135, True)

    bcc_noised_165 = rotate_image(bcc_noised, 165)
    bcc_noised_165_cw = rotate_image(bcc_noised, 165, True)

    bcc_noised_210 = rotate_image(bcc_noised, 210)
    bcc_noised_210_cw = rotate_image(bcc_noised, 210, True)
    
    bcc_noised_rotated = bcc_noised_15 + bcc_noised_30 + bcc_noised_45 + bcc_noised_135 + bcc_noised_165 + bcc_noised_210 + bcc_noised_15_cw + bcc_noised_30_cw + bcc_noised_45_cw + bcc_noised_135_cw + bcc_noised_165_cw + bcc_noised_210_cw
    return random.sample(bcc_noised_rotated, 514)

bcc_noised_rotated = rotate_noised_bcc_images(bcc_noised) # 1


# flip + rotate
def rotate_flipped_bcc_images(bcc_flipped):
    bcc_flipped_15 = rotate_image(bcc_flipped, 15)
    bcc_flipped_15_cw = rotate_image(bcc_flipped, 15, True)

    
    bcc_flipped_30 = rotate_image(bcc_flipped, 30)
    bcc_flipped_30_cw = rotate_image(bcc_flipped, 30, True)
    
    bcc_flipped_45 = rotate_image(bcc_flipped, 45)
    bcc_flipped_45_cw = rotate_image(bcc_flipped, 45, True)

    bcc_flipped_60 = rotate_image(bcc_flipped, 60)
    bcc_flipped_60_cw = rotate_image(bcc_flipped, 60, True)

    bcc_flipped_150 = rotate_image(bcc_flipped, 150)
    bcc_flipped_150_cw = rotate_image(bcc_flipped, 150, True)

    bcc_flipped_195 = rotate_image(bcc_flipped, 195)
    bcc_flipped_195_cw = rotate_image(bcc_flipped, 195, True)

    bcc_flipped_210 = rotate_image(bcc_flipped, 120)
    bcc_flipped_210_cw = rotate_image(bcc_flipped, 10, True)
    
    bcc_flipped_rotated = bcc_flipped_15 + bcc_flipped_30 + bcc_flipped_45 + bcc_flipped_60 + bcc_flipped_150 + bcc_flipped_195 + bcc_flipped_210 + bcc_flipped_15_cw + bcc_flipped_30_cw + bcc_flipped_45_cw + bcc_flipped_60_cw + bcc_flipped_150_cw + bcc_flipped_195_cw + bcc_flipped_210_cw
    return random.sample(bcc_flipped_rotated, 771)
    
bcc_flipped_rotated = rotate_flipped_bcc_images(bcc_flipped) # 1.5
```

```python
# noise + flip + rotate
def rotate_noised_flipped_bcc_images(bcc_noised_flipped):
    bcc_noised_flipped_15 = rotate_image(bcc_noised_flipped, 15)
    bcc_noised_flipped_15_cw = rotate_image(bcc_noised_flipped, 15, True)

    bcc_noised_flipped_30 = rotate_image(bcc_noised_flipped, 30)
    bcc_noised_flipped_30_cw = rotate_image(bcc_noised_flipped, 30, True)

    bcc_noised_flipped_45 = rotate_image(bcc_noised_flipped, 45)
    bcc_noised_flipped_45_cw = rotate_image(bcc_noised_flipped, 45, True)

    bcc_noised_flipped_60 = rotate_image(bcc_noised_flipped, 60)
    bcc_noised_flipped_60_cw = rotate_image(bcc_noised_flipped, 60, True)

    bcc_noised_flipped_120 = rotate_image(bcc_noised_flipped, 120)
    bcc_noised_flipped_120_cw = rotate_image(bcc_noised_flipped, 120, True)

    bcc_noised_flipped_135 = rotate_image(bcc_noised_flipped, 135)
    bcc_noised_flipped_135_cw = rotate_image(bcc_noised_flipped, 135, True)

    bcc_noised_flipped_195 = rotate_image(bcc_noised_flipped, 195)
    bcc_noised_flipped_195_cw = rotate_image(bcc_noised_flipped, 195, True)

    bcc_noised_flipped_rotated = bcc_noised_flipped_15 + bcc_noised_flipped_30 + bcc_noised_flipped_45 + bcc_noised_flipped_60 + bcc_noised_flipped_120 + bcc_noised_flipped_135 + bcc_noised_flipped_195 + bcc_noised_flipped_15_cw + bcc_noised_flipped_30_cw + bcc_noised_flipped_45_cw + bcc_noised_flipped_60_cw + bcc_noised_flipped_120_cw + bcc_noised_flipped_135_cw + bcc_noised_flipped_195_cw

    return random.sample(bcc_noised_flipped_rotated, 771)

bcc_noised_flipped_rotated = rotate_noised_flipped_bcc_images(bcc_noised_flipped)
```

```python
ind_bcc = bcc_images + bcc_noised + bcc_flipped + bcc_rotated + bcc_noised_flipped + bcc_noised_rotated + bcc_flipped_rotated + bcc_noised_flipped_rotated
```
